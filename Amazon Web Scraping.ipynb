# Checking Whether any change in Price

```python
#Checking Whether any change in price
import os
import logging

# Setup logging
logging.basicConfig(filename='scraper.log', level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

URL = 'https://www.amazon.in/Titan-Karishma-Stainless-Watch-NL1639SM02-NN1639SM02/dp/B00ISNVQMW/'
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/137.0.0.0 Safari/537.36"
}
CSV_FILE = 'AmazonWebScraper.csv'
TARGET_PRICE = 1000  # Set your desired price to be notified

def create_csv_if_not_exists():
    if not os.path.exists(CSV_FILE):
        with open(CSV_FILE, 'w', newline='', encoding='UTF8') as f:
            writer = csv.writer(f)
            writer.writerow(['Title', 'Price (Rs)', 'Date'])

def get_product_info():
    try:
        page = requests.get(URL, headers=HEADERS)
        soup = BeautifulSoup(page.content, "html.parser")
        title = soup.find(id='productTitle').get_text().strip()
        price = soup.find('span', class_='a-price-whole').get_text().strip().replace(',', '')

        # Convert price to integer
        price = int(price)

        return title, price
    except Exception as e:
        logging.error(f"Error fetching product info: {e}")
        return None, None

def check_price_and_store():
    create_csv_if_not_exists()
    title, price = get_product_info()

    if title and price:
        today = datetime.date.today()

        with open(CSV_FILE, 'a+', newline='', encoding='UTF8') as f:
            writer = csv.writer(f)
            writer.writerow([title, price, today])

        logging.info(f"Recorded: {title} at ₹{price} on {today}")

        # Alert condition
        if price < TARGET_PRICE:
            print(f"Price drop alert! The product is now ₹{price}")
        else:
            print(f"No alert. Current price: ₹{price}")
    else:
        print("Failed to retrieve product info.")
if __name__ == "__main__":
    check_price_and_store()
```
