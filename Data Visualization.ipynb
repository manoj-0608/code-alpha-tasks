# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading the Dataset

```python
df=pd.read_csv('retail_sales.csv')
```

# Set Style

```python
sns.set(style='whitegrid')
```

# 1.Distribution Plots

```python
plt.figure(figsize=(12,8))
plt.subplot(1, 2, 1)
sns.histplot(df['Units_Sold'], bins=15, kde=True, color='green')
plt.title('Units Sold Distribution')
plt.subplot(1, 2, 2)
sns.histplot(df['Total_Sales'], bins=15, kde=True, color='red')
plt.title('Total Sales Distribution')
plt.tight_layout()
plt.show()
```

# 2.Correlation Heatmap

```python
plt.figure(figsize=(10, 8))
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
corr_matrix = df[numeric_cols].corr()
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    linewidths=0.5,
    square=True,
    cbar_kws={"shrink": 0.8},
    annot_kws={"size": 10}
)
plt.title('Correlation Heatmap of Retail Numeric Data', fontsize=14)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

# 3.Scatter Plot=Units Sales vs Total Sales

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='Units_Sold',
    y='Total_Sales',
    hue='Product_Category',
    data=df,
    s=100,
    alpha=0.7
)
plt.title('Units Sold vs Total Sales (100 Transactions)')
plt.xlabel('Units Sold')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

# 4.Pie Chart = Payment Method Distribution

```python
payment_counts = df['Payment_Method'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Payment Method Distribution')
plt.axis('equal')  
plt.show()
```

# 5.Line Chart = Total Sales Over Time

```python
df['Date'] = pd.to_datetime(df['Date'])  # Convert Date column to datetime
daily_sales = df.groupby('Date')['Total_Sales'].sum().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(x='Date', y='Total_Sales', data=daily_sales)
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

# 6.Violin Plot = Units Sold By Product Category

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x='Product_Category', y='Units_Sold', data=df)
plt.title('Units Sold by Product Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
